```python
import seaborn as sns
import ete3
import glob
import numpy as np
iqs = glob.glob( './examples/*.treefile' )


fast = glob.glob('./examples/*.fasttree')
print(iqs)

haps = [ f for f in fast + iqs if 'hap' in f ]+['./hapdistmat.txtfastmemat.txt_fastme_tree.nwk']
wnts = [ f for f in fast + iqs if 'wnt' in f ]+['./wntdistmat.txtfastmemat.txt_fastme_tree.nwk']

htrees = { file:ete3.Tree( file) for file in haps }
wntrees = { file:ete3.Tree( file) for file in wnts }

print(wntrees['./wntdistmat.txtfastmemat.txt_fastme_tree.nwk'])

for haptrees in [htrees, wntrees]:
    treedists = np.zeros((len(haptrees),len(haptrees)))
    for i,t1 in enumerate(haptrees):
        for j,t2 in enumerate(haptrees):
            if i < j:
                res= haptrees[t1].robinson_foulds(haptrees[t2] , unrooted_trees = True) 
                rf, maxrf = res[0:2]
                treedists[i,j] = rf/maxrf
    treedists+=treedists.T

    sns.clustermap( treedists , xticklabels = list(haptrees.keys()) , yticklabels = list(haptrees.keys()) )
```
